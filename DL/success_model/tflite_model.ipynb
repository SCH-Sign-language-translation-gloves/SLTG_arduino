```python
from tensorflow import keras
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../data/final_data.csv')
df
```

```python
np.random.seed(42)
dfs = df.copy()

# shuffling 하고 index reset
dfs = dfs.sample(frac = 1).reset_index(drop=True)
dfs
```

```python
# X,y 나누기
data = dfs.copy()
X = data.drop("number", axis =1)
X
```

```python
y = dfs["number"].copy()
y
```

```python
x_test, x_train = X[:15000], X[15000:]
y_test, y_train = y[:15000], y[15000:]
x_train
```

```python
tf.random.set_seed(42)
np.random.seed(42)

model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[11,]),
    keras.layers.Dense(300, kernel_initializer="lecun_normal", activation = 'selu'),
    keras.layers.Dense(300, kernel_initializer="lecun_normal", activation = 'selu'),
    keras.layers.Dense(31, activation="softmax")
])
```

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer='adam',
              metrics=["accuracy"])
```

```python
lr_schedular = keras.callbacks.ReduceLROnPlateau(patience=10, factor = 0.5)
```

```python
history = model.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test), callbacks = [lr_schedular])
```

```python
val_loss, val_accuracy = model.evaluate(x_test, y_test)
print(val_accuracy)
```

```python
model.save("model_0524.h5")
```

```python
model = tf.keras.models.load_model('model_0524.h5')
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
```

```python
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS,
                                       tf.lite.OpsSet.SELECT_TF_OPS]
converter.target_spec.supported_types = [tf.float16]
converter.optimizations = [tf.lite.Optimize.DEFAULT]
```

```python
tflite_model = converter.convert() 
```

```python
with open('model0524.tflite', 'wb') as f:
    f.write(tflite_model)
```
