```python
from tensorflow import keras
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../data/final_data.csv')
df
```

```python
np.random.seed(42)
dfs = df.copy()

# shuffling í•˜ê³  index reset
dfs = dfs.sample(frac = 1).reset_index(drop=True)
dfs
```

```python
# X,y ë‚˜ëˆ„ê¸°
data = dfs.copy()
X = data.drop("number", axis =1)
X
```

```python
y = dfs["number"].copy()
y
```

```python
X
```

```python
y
```

```python
tf.random.set_seed(42)
np.random.seed(42)

model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[11,]),
    keras.layers.Dense(300, kernel_initializer="lecun_normal", activation = 'selu'),
    keras.layers.Dense(300, kernel_initializer="lecun_normal", activation = 'selu'),
    keras.layers.Dense(31, activation="softmax")
])
```

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer='adam',
              metrics=["accuracy"])
```

```python
lr_schedular = keras.callbacks.ReduceLROnPlateau(patience=10, factor = 0.5)
```

```python
history = model.fit(X, y, epochs=100)
```

```python
model.save("model_0525.h5")
```
